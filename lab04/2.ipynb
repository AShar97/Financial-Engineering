```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from backtester.dataSource.yahoo_data_source import YahooStockDataSource

startDateStr = '2016/12/31'
endDateStr = '2017/12/31'
cachedFolderName = '/Users/ayushsharma/Documents/Finance/yahooData/'
dataSetId = 'testPairsTrading'
instrumentIds = ['GOOGL','FB','AAPL','MSFT','AMZN','ADBE','TYO','XOM','BP','WMT']
#instrumentIds = ['RDSA']
ds = YahooStockDataSource(cachedFolderName=cachedFolderName,
                            dataSetId=dataSetId,
                            instrumentIds=instrumentIds,
                            startDateStr=startDateStr,
                            endDateStr=endDateStr,
                            event='history')
stock_prices = ds.getBookDataByFeature()['Adj Close']
```

```python
returns = ((stock_prices.shift(-1) - stock_prices)/stock_prices).dropna()
returns
```

```python
Cov = returns.cov() * 252
Cov
```

```python
Mean = returns.mean() * 252
Mean
```

```python
Vol = returns.std() * np.sqrt(252)
Vol
```

```python
C = Cov.values
C
```

```python
M = Mean.values.reshape(Mean.shape[0],1)
M
```

```python
unity = np.ones(M.shape[0]).reshape(M.shape[0],1)
```

```python
M.shape
```

```python
mu_v = np.arange(-1,2, 0.001)
mu_v = mu_v.reshape(mu_v.shape[0], 1)
```

```python
a = np.dot(np.dot(unity.T,np.linalg.inv(C)),unity)[0,0]
b = np.dot(np.dot(unity.T,np.linalg.inv(C)),M)[0,0]
c = np.dot(np.dot(M.T,np.linalg.inv(C)),M)[0,0]
Del = (a*c) - (b**2)

var = (((a*(mu_v**2)) - (2*b*mu_v) + c)/Del)

sigma_v = np.sqrt(var)
```

```python
rf = 0.07
gammaW_m = np.dot((M - rf*unity).T,np.linalg.inv(C))
W_m = gammaW_m/np.sum(gammaW_m)
W_m
```

```python
mu_m = np.dot(W_m, M)[0,0]
sigma_m = np.sqrt(np.dot(np.dot(W_m, C),W_m.T))[0,0]
```

```python
sigma_cml = np.arange(0, 0.5, 0.001)
sigma_cml = sigma_cml.reshape(sigma_cml.shape[0],1)

mu_cml = rf + ((mu_m-rf)/sigma_m)*sigma_cml
```

```python
plt.plot(np.sqrt(np.diag(C)),M, '*')
plt.plot(sigma_v, mu_v)
plt.plot(sigma_m,mu_m, '*')
plt.plot(0,rf, '*')
plt.plot(sigma_cml,mu_cml)
plt.title("Portfolio Management")
plt.xlabel("Risk (Standard Deviation)")
plt.ylabel("Return (Mean)")

plt.legend(["Assets","Markowitz efficient frontier", "Market portfolio", "Risk-free security", "Capital Market Line"])
plt.show()
```

```python
beta = (M - rf)/(mu_m - rf)
```

```python
beta_sml = np.arange(-0.25, 1.5, 0.001)
mu_sml = rf + (mu_m - rf)*beta_sml
```
